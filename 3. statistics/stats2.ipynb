### 1.What is the meaning of six sigma in statistics?  Give proper example
Six Sigma is about making sure your processes are so good that mistakes are extremely rare. It's like aiming for perfection in everything you do!

Imagine you're making sandwiches. You want to make sure every sandwich you make is as perfect as possible. Six Sigma is like having a super strict quality control system to ensure your sandwiches turn out amazing every time.

In Six Sigma, you measure how many mistakes or defects you have in making sandwiches. Let's say you aim for only one mistake in every million sandwiches you make. That's a really high standard!

So, if you make a million sandwiches and only one has a mistake, you're operating at Six Sigma. It means your process is incredibly precise and consistent.

### 3.What is the meaning of the five-number summary in Statistics? Give proper example

Minimum: The smallest value in the dataset.

First Quartile (Q1): The value below which 25% of the data falls. It is also known as the lower quartile.

Median (Q2): The middle value of the dataset when arranged in ascending order. It is also the value below which 50% of the data falls. If the dataset has an odd number of observations, the median is the middle value. If the dataset has an even number of observations, the median is the average of the two middle values.

Third Quartile (Q3): The value below which 75% of the data falls. It is also known as the upper quartile.

Maximum: The largest value in the dataset.


Consider the following dataset: 12, 15, 17, 20, 22, 25, 28, 30, 33, 36

Minimum: 12

Q1 (First Quartile): 17 (Value below which 25% of the data falls)

Median (Q2): 22 (Middle value of the dataset)

Q3 (Third Quartile): 30 (Value below which 75% of the data falls)

Maximum: 36

### 4.What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook

Correlation measures the strength and direction of the relationship between two variables. It indicates how much one variable changes when the other variable changes. Correlation values range from -1 to 1:

If the correlation value is close to 1, it indicates a strong positive correlation, meaning that as one variable increases, the other variable tends to increase as well.

If the correlation value is close to -1, it indicates a strong negative correlation, meaning that as one variable increases, the other variable tends to decrease.

If the correlation value is close to 0, it indicates little to no linear relationship between the variables.

#### we have a dataset containing the number of hours studied and the corresponding exam scores for a group of students:

```python
import pandas as pd

# Create a DataFrame from the dataset
data = {'Hours Studied': [2, 3, 4, 5, 6],
        'Exam Score': [60, 70, 75, 80, 85]}
df = pd.DataFrame(data)

# Calculate the correlation
correlation = df['Hours Studied'].corr(df['Exam Score'])
print("Correlation coefficient:", correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='Hours Studied', y='Exam Score', data=df, scatter_kws={'s': 100}, line_kws={'color': 'red'})
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.show()
```
